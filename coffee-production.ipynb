```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

coffee = pd.read_csv("Coffee_production.csv")
print(coffee.head())
```

```python
print(coffee.columns)
```

```python
print(coffee['Coffee type'].value_counts())
print(coffee['Country'].count())
```

```python
plot1 = sns.relplot(data=coffee, x="Country", y="Total_production", hue="Country", size="Total_production", legend=False)
plot1.set(yscale="log")
plot1.set(title="Total Production for each Country")
plt.xticks(rotation=90, size=5)
plt.show()
plot1.savefig('total-production.png')
```

```python
plot2 = sns.countplot(data=coffee, x="Coffee type", hue="Coffee type", palette="muted")
plot2.set(title="Countries categorized by Coffee Type")
plt.show()
plt.savefig('coffee-types')
```

```python
arabica = coffee[coffee["Coffee type"] == "Arabica"][['Country','Total_production']]
print(arabica)
plot3 = sns.barplot(data=arabica, x="Country", y="Total_production", hue="Country")
plot3.set_yscale("log")
plot3.set(title="Arabica Production")
plt.xticks(rotation=90)
plt.show()
plt.savefig('arabica-production')
```

```python
robusta = coffee[coffee["Coffee type"] == "Robusta"][['Country','Total_production']]
print(robusta)
plot4 = sns.barplot(data=robusta, x="Country", y="Total_production", hue="Country")
plot4.set_yscale("log")
plot4.set(title="Robusta Production")
plt.xticks(rotation=90)
plt.show()
plt.savefig('robusta-production')
```

```python
colombia = coffee[coffee['Country'] == "Colombia"]
plot5 = sns.barplot(data=colombia)
plot5.set(yscale="log")
plot5.set(title="Colombia Production")
plt.xticks(rotation=90, size=5)
plt.show()
plt.savefig('colombia-production')
```

```python
cotedIvoire = coffee[coffee['Country'] == "Cte d'Ivoire"]
plot6 = sns.barplot(data=cotedIvoire)
plot6.set(yscale="log")
plot6.set(title="Cte d'Ivoire Production")
plt.xticks(rotation=90, size=5)
plt.show()
plt.savefig('cotedIvoire-production')
```
